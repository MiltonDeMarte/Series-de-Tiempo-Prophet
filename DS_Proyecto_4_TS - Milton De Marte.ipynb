## **Pre-procesamiento 2017-2018-2019**

```python
data_2019 = pd.read_csv('Datasets/flujo-vehicular-2019.csv', sep =',')
data_2018 = pd.read_csv('Datasets/flujo-vehicular-2018.csv', sep =',')
data_2017 = pd.read_csv('Datasets/flujo-vehicular-2017.csv', sep =',')
```

```python
dataset = pd.concat([data_2019, data_2018, data_2017])


mask = np.logical_and(dataset.estacion=='Illia', dataset.forma_pago == 'EFECTIVO')
dataset = dataset[mask]

dataset.drop(columns = ['periodo','hora_inicio','forma_pago', 'estacion'], inplace = True)

dataset.head()
```

```python
dataset['fecha2'] = pd.to_datetime(dataset.fecha) + pd.to_timedelta(dataset.hora_fin, unit = 'h')

dataset.drop(columns=['fecha','hora_fin'], inplace = True)

dataset.head()
```

```python
dataset.rename(columns = {'fecha2':'fecha', "cantidad_pasos":"cantidad_autos"}, inplace = True) 
dataset.head()
```

```python
dataset.sort_values('fecha', inplace = True)
dataset.reset_index(drop=True, inplace = True)
dataset.head()
dataset.shape
```

```python
dataset.describe()
```

```python
plt.figure(figsize = (15,6))
plt.plot(dataset.fecha, dataset.cantidad_autos)
plt.title("Flujo de autos medido en Horas \n", color= 'r', fontsize= 15)
plt.xlabel('\n Tiempo', color= 'g', fontsize= 15)
plt.ylabel('\n Cantidad de autos, por hora', color= 'g', fontsize= 15)
plt.show()
```

```python
diario = dataset.resample('D', on = 'fecha').sum()
print(diario.shape)
diario.head()
```

```python
diario.describe()
```

```python
plt.figure(figsize = (15,6))

plt.plot(diario.index, diario.cantidad_autos)
plt.title("Flujo de autos medido en Dias \n", color= 'r', fontsize= 15)
plt.xlabel('\n Tiempo', color= 'g', fontsize= 15)
plt.ylabel('\n Cantidad de autos, por dia', color= 'g', fontsize= 15)

plt.show()
```

### *Interpolando Valores Atipicos*

```python
diario[(diario.cantidad_autos == 0) | (diario.cantidad_autos <= 10000)] = np.NaN

print(diario.shape)

print(diario.isna().sum())

diario= diario.interpolate(method='nearest', k=4)

print(diario.isna().sum())
```

```python
plt.figure(figsize = (15,6))

plt.plot(diario.index, diario.cantidad_autos,  color= "crimson")
plt.title("Flujo de vehiculos frecuencia Diaria (sin Valores Atipicos)", color= 'indianred', fontsize= 20)
plt.xlabel('\n Tiempo', color= 'indigo', fontsize= 15)
plt.ylabel('\n Cantidad de autos, por dia', color= 'indigo', fontsize= 15, loc='top', rotation=0)
plt.ylim((0,80000))

plt.show()
```

## **S.A.R.I.M.A.X**

```python
y= semanal.cantidad_autos


model= sm.tsa.statespace.SARIMAX(y,
                                order=(10, 1, 8),
                                seasonal_order=(10, 1, 8, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results= model.fit()

print(results.summary().tables[1])
```

### *Prediccion 4to Trimestre 2019*

* _*Vuelvo a generar la prediccion de SARIMAX para el ultimo trimestre del año 2019 para luego, comparar con las prediccion de Prophet y LSTM sobre el mismo tiempo observado. Confrontar los resultados y observar si los nuevos modelos presentan mejoras sustanciables para con SARIMAX.*_


**Observacion personal:**

* _*Por mi parte lo veo "dificil" que los nuevos modelos mejoren la prediccion de SARIMAX para ultimo trimestre de 2019. El mismo devolvio, no solo una gran interpretacion y adaptacion a los valores de la serie sino tambien, un excelente resultado luego de ser evaluado a traves de la metrica elegida(RMSE). Se esperan obtener mejores resultados aunque se anticipa la no-obtencion de los mismos. Continuo con la observacion a lo largo del trabajo.*_

```python
prediction = results.get_prediction(start= dt.datetime(2019,10,6), dynamic=False)
pred_conf = prediction.conf_int()

plt.figure(figsize=(20,10))

ax = y['2017':].plot(label='Serie Observada')
prediction.predicted_mean.plot(ax=ax, label='SARIMAX One-Step-Prediction', alpha=0.7)

ax.fill_between(pred_conf.index,
                pred_conf.iloc[:, 0],
                pred_conf.iloc[:, 1], color='k', alpha=0.2)

plt.title("Desempeño SARIMAX One-Step Trimestre-2019", color= 'k', fontsize=25)
ax.set_xlabel("\n Fecha", fontsize=18, color='indigo')
ax.set_ylabel("\n  Cantidad Autos", loc= 'top', rotation=0, fontsize=18,color='indigo')
plt.xlim(dt.datetime(2019,1,1), dt.datetime(2019,12,29))
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)

plt.legend(fontsize=15)
plt.show()

y_forecast= prediction.predicted_mean
y_real = y['2019-10-06':]


rmse_sarimax_one = np.sqrt(metrics.mean_squared_error(y_forecast, y_real))
print("RMSE- SARIMAX: ", rmse_sarimax_one)
```

### *Hypertunning*

```python
def gs_prophet(params, model, df, rolling_window, horizon):
    '''Brute-Force GridSearch for Prophet models
    Rolling-Window must be an int
    Horizon must be denoted as: 'X Y'
         Where X is equal to a number 
               Y is equal to frequency(days or less) - E.G = '30 days' '''
    rmse_list = []  
    for i in params:
        m = model(**i).fit(df) 
        df_cv = cross_validation(m, horizon= horizon, parallel="processes")
        df_p = performance_metrics(df_cv, rolling_window=rolling_window)
        rmse_list.append(df_p['rmse'].values[0])
    tuning_results = pd.DataFrame(params)
    tuning_results['rmse'] = rmse_list
    return tuning_results, tuning_results.loc[tuning_results['rmse'].idxmin()]
```

#### *Creating Params*

```python
param_grid = {  
    'changepoint_prior_scale': np.arange(0.001, 1, .08),
    'seasonality_prior_scale': np.arange(0.01, 10, 0.5)}

all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
```

##### ACLARACION IMPORTANTE

* _*Esta combinaciones de distintos paramaetros para Prophet estuvo corriendo por mas de 5 horas. Si bien no es un impedimiento, recomiendo no correr de nuevo esta celda ni las 2 celdas de codigo posteriores(dado que estan sujetas a esta linea de codigo) para no retrarse en la ejecucion del NB.*_
